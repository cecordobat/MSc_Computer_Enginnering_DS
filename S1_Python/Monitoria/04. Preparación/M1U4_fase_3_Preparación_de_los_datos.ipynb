```python
clean_clinal_dset['menopause_status'] = clean_clinal_dset['menopause_status'].replace('Peri (6-12 months since last menstrual period)',
                                                                                      'Pre (<6 months since LMP AND no prior bilateral ovariectomy AND not on estrogen replacement)')

clean_clinal_dset['pathologic_T'] = clean_clinal_dset['pathologic_T'].map(lambda x: x[:2])
clean_clinal_dset['pathologic_N'] = clean_clinal_dset['pathologic_N'].map(lambda x: x[:2])
clean_clinal_dset['pathologic_M'] = clean_clinal_dset['pathologic_M'].map(lambda x: x[x.find('M'):x.find('M')+2])

clean_clinal_dset['pathologic_stage'] = clean_clinal_dset['pathologic_stage'].map(lambda x: str(x).replace('A','').replace('B','').replace('C',''))
clean_clinal_dset.head()
```

Considerar la resolución de los factores pronosticos.

### **1.5. Datos corruptos**
---
En la detección de datos corruptos, pregunte si hay posibles errores de entrada o problemas de transferencia que podrían haber afectado la integridad de los datos.
* ¿Cómo se determinó la presencia de datos corruptos en el conjunto de datos?
* ¿Qué medidas se tomaron para corregir o eliminar los datos corruptos?
* ¿Hubo casos en los que la corrupción de datos no pudo ser manejada y cómo se justificó esa decisión?

```python
clean_clinal_dset['tissue_prospective_collection_indicator'] = clean_clinal_dset['tissue_prospective_collection_indicator'].fillna('NO')

clean_clinal_dset['radiation_therapy'] = clean_clinal_dset['radiation_therapy'].fillna('YES')

clean_clinal_dset['lymph_node_examined_count'] = clean_clinal_dset['lymph_node_examined_count'].fillna(np.median(clean_clinal_dset['lymph_node_examined_count'].dropna()))

clean_clinal_dset['breast_carcinoma_surgical_procedure_name'] = clean_clinal_dset['breast_carcinoma_surgical_procedure_name'].fillna('Modified Radical Mastectomy')

clean_clinal_dset['menopause_status'] = clean_clinal_dset['menopause_status'].fillna('Post (prior bilateral ovariectomy OR >12 mo since LMP with no prior hysterectomy)')

clean_clinal_dset['pathologic_stage'] = clean_clinal_dset['pathologic_stage'].fillna('Stage X')
clean_clinal_dset['pathologic_stage'] = clean_clinal_dset['pathologic_stage'].replace('nan','Stage X')

clean_clinal_dset.head()
```

```python
clean_clinal_dset = clean_clinal_dset[(clean_clinal_dset['days_to_last_followup'] >= 365) | (clean_clinal_dset['days_to_last_followup'].isna())]
clean_clinal_dset.head()
```

Eliminados 154 regsitros, restantes 911.

```python
clean_clinal_dset.to_csv('clean_clinical.csv', index = False)
prof = ProfileReport(clean_clinal_dset, minimal=False)
prof.to_file(output_file='./clean adapted clinical data.html')
```
