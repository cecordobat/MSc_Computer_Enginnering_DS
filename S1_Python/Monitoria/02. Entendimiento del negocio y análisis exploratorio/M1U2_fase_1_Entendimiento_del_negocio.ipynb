El cáncer de seno es una de las enfermedades más prevalentes a nivel mundial, sin embargo, esta pude categorizarse en cuatro subtipos: Luminal A, Luminal B, HER 2 y triple negativo (TNBC). Este es la segunda categoría más prevalente a nivel mundial (20%) y la más mortal (30%).

Para combatir la enfermedad se hace uso de quimioterapia convencional, este es un tratamiento supremamente agresivo, generando múltiples efectos secundarios que perduran a largo plazo. La clave para la formulación de tratamientos efectivos e inocuos es el entendimiento y pronostico personalizado de cada caso, lo cual depende de una correcta estimación del tiempo del paciente.

En proyecto titulado "Modelo para la estimación de supervivencia en cáncer de seno triple negativo" se prende construir un modelo de estimación de supervivencia a partir de la información a nivel de paciente. Permitiendo a los especialistas entender la severidad de cada caso de manera individual.

Dadas la característica del problema, se plantea la implementación de una regresión de Cox, que emplee como covariables la  información clínica del paciente (edad, raza, patrones de comportamiento y características morfo-fisiologicas del tumor).

Si bien se espera que el modelo implementado apoye la decisión de un especialista, su implementación tiene algunos limitantes:

* El modelo propuesto estará diseñado para entender la diferencia entre la severidad de cada caso, el tiempo estimado es un parámetro altamente afectado para la población de estudio.
* El modelo realizará inferencias, a partir de la totalidad de la descripción de un paciente, por lo que deberán ser entregadas la totalidad de las covariable estudiadas para cada caso.
* El modelo será un apoyo para el especialista, pero bajo ningún concepto remplaza la opinión médica.
